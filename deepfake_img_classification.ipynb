```python
import cv2
import numpy as np

def preprocess_and_save_images(csv_file, output_folder, image_size=(128, 128), batch_size=1000):
    df = pd.read_csv(csv_file)
    os.makedirs(output_folder, exist_ok=True)  
    images, labels = [], []
    batch_count = 0
    
    for idx, row in df.iterrows():
        image_path = row['image_path']
        label = row['label']
        image = cv2.imread(image_path)
        if image is None:
            print(f"Warning: Unable to read image at {image_path}")
            continue
        image_resized = cv2.resize(image, image_size)
        image_normalized = image_resized / 255.0
        images.append(image_normalized)
        labels.append(label)
        if len(images) == batch_size:
            save_batch(images, labels, output_folder, batch_count)
            batch_count += 1
            images, labels = [], []  
    
    if images:
        save_batch(images, labels, output_folder, batch_count)
        print(f"Saved final batch: {batch_count}")

    print("Preprocessing complete!")

def save_batch(images, labels, output_folder, batch_count):
    images = np.array(images, dtype=np.float32)
    labels = np.array(labels, dtype=np.int32)
    
    np.save(os.path.join(output_folder, f"images_batch_{batch_count}.npy"), images)  
    np.save(os.path.join(output_folder, f"labels_batch_{batch_count}.npy"), labels)  
    print(f"Saved batch {batch_count} with {len(images)} images.")

csv_file = 'image_dataset.csv'
output_folder = 'preprocessed_data'
image_size = (128, 128)  
batch_size = 1000  

preprocess_and_save_images(csv_file, output_folder, image_size, batch_size)
```

```python
import numpy as np
preprocessed_folder = 'preprocessed_data'

def load_batches(folder):
    image_batches = sorted([os.path.join(folder, f) for f in os.listdir(folder) if f.startswith('images_batch')])
    label_batches = sorted([os.path.join(folder, f) for f in os.listdir(folder) if f.startswith('labels_batch')])
    for img_file, lbl_file in zip(image_batches, label_batches):
        images = np.load(img_file)
        labels = np.load(lbl_file)
        yield images, labels

for images, labels in load_batches(preprocessed_folder):
    print(f"Batch shape: {images.shape}, Labels shape: {labels.shape}")
    break
```

```python
from sklearn.model_selection import train_test_split

image_batches = sorted([os.path.join(preprocessed_folder, f) for f in os.listdir(preprocessed_folder) if f.startswith('images_batch')])
label_batches = sorted([os.path.join(preprocessed_folder, f) for f in os.listdir(preprocessed_folder) if f.startswith('labels_batch')])

train_images, temp_images, train_labels, temp_labels = train_test_split(image_batches, label_batches, test_size=0.3, random_state=42)
val_images, test_images, val_labels, test_labels = train_test_split(temp_images, temp_labels, test_size=0.33, random_state=42)

def load_batch_subset(image_files, label_files):
  for img_file, lbl_file in zip(image_files, label_files):
    images = np.load(img_file)
    labels = np.load(lbl_file)
    yield images, labels
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout,Input

model = Sequential([
    Input(shape=(128, 128, 3)),  
    
    Conv2D(32, (3, 3), activation='relu', padding='same'),
    MaxPooling2D(pool_size=(2, 2)),
    
    Conv2D(64, (3, 3), activation='relu', padding='same'),
    MaxPooling2D(pool_size=(2, 2)),
    
    Conv2D(128, (3, 3), activation='relu', padding='same'),
    MaxPooling2D(pool_size=(2, 2)),
    
    Flatten(),
    
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')  
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
for epoch in range(10):
    for train_images_batch, train_labels_batch in load_batch_subset(train_images, train_labels):
        model.fit(train_images_batch, train_labels_batch, epochs=1, verbose=1)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt

def load_test_data(image_files, label_files):
    all_images = []
    all_labels = []
    
    for img_file, lbl_file in zip(image_files, label_files):
        images = np.load(img_file)
        labels = np.load(lbl_file)
        all_images.append(images)
        all_labels.append(labels)
    
    return np.concatenate(all_images), np.concatenate(all_labels)

test_images, test_labels = load_test_data(test_images, test_labels)

test_loss, test_accuracy = model.evaluate(test_images, test_labels)

print(f"Test Loss: {test_loss:.4f}")
print(f"Test Accuracy: {test_accuracy:.4f}")

test_preds = model.predict(test_images)

predicted_classes = (test_preds > 0.5).astype(int).flatten() 

print("\nClassification Report:")
print(classification_report(test_labels, predicted_classes))
```
